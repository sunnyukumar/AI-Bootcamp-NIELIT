```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 1. Generate synthetic data for demonstration
# make_blobs creates isotropic Gaussian blobs for clustering
X, _ = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=0.60)

# 2. Initialize and fit the K-Means model
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto')
kmeans.fit(X)

# 3. Get cluster assignments and centroids
labels = kmeans.labels_  # Cluster labels for each data point
centroids = kmeans.cluster_centers_  # Coordinates of the cluster centroids

# 4. Visualize the clusters and centroids
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.8)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, label='Centroids')

plt.title("K-Means Clustering")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

print("Cluster Centroids:\n", centroids)
```
